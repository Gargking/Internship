# World Happiness Report

__Context__

The World Happiness Report is a landmark survey of the state of global happiness. The first report was published in 2012, the second in 2013, the third in 2015, and the fourth in the 2016 Update. The World Happiness 2017, which ranks 155 countries by their happiness levels, was released at the United Nations at an event celebrating International Day of Happiness on March 20th. The report continues to gain global recognition as governments, organizations and civil society increasingly use happiness indicators to inform their policy-making decisions. Leading experts across fields – economics, psychology, survey analysis, national statistics, health, public policy and more – describe how measurements of well-being can be used effectively to assess the progress of nations. The reports review the state of happiness in the world today and show how the new science of happiness explains personal and national variations in happiness.

We need to check again by using __Regularization__ 

To regularize the model, a Shrinkage penalty is added to the cost function. Let's see different types of regularizations in regression.

• LASSO

• RIDGE

• ELASTICNET(Less popular)

```python
from sklearn.linear_model import Ridge, Lasso, RidgeCV, LassoCV

# LassoCV will return best alpha after max iteration
# Normalize is substracting the mean and dividing by the L2-norm
lasscv = LassoCV(alphas = None, max_iter = 100, normalize = True) # if dataset is big INC max_iter
lasscv.fit(x_train, y_train)

# best alpha paremeter
alpha = lasscv.alpha_
alpha

# now that we have best paremeter, Let's use Lasso regression and see how well our data has fitted before

lasso_reg = Lasso(alpha)
lasso_reg.fit(x_train, y_train)

lasso_reg.score(x_test, y_test)
```

```python
# RidgeCV will return best alpha and coefficients after performing 10 cross validation.

ridgecv = RidgeCV(alphas = np.arange(0.001,0.1,0.01), normalize = True)

ridgecv.fit(x_train, y_train)

ridgecv.alpha_

ridge_model = Ridge(alpha=ridgecv.alpha_)
ridge_model.fit(x_train, y_train)

ridge_model.score(x_test, y_test)
```

We got around same R2 square using Ridge regression as well. So, its safe to say there is no overfitting.

So, we can see by using different type of regularization, we still are getting the same R2 score. That means our OLS model has been well trained over the training data and there is no overfitting.
